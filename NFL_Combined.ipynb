```python
# Dependencies
import pandas as pd
import os

#Setting Datasets
DL2010 = "Resources/QB_2010_to_2015.csv"
#Loading into DataFrame
DL_2010_df = pd.read_csv(DL2010, encoding="utf-8")
DL_2010_df.head()
```

```python
# Creating a new csv that combines OLB with ILB
output = "Output/Combined_2010_2015.csv"

df = pd.DataFrame()

filenames = os.listdir( "./Resources" )  #["OLB_2010_to_2015.csv", "ILB_2010_to_2015.csv"]

for filename in filenames:
    df = df.append(pd.read_csv("./Resources/" + filename), ignore_index=True)
    
#Replacing names of positions for easier data analysis
#Outside Linebacker & Inside Linebacker converted to Linebacker allocation
df['Pos'] = df['Pos'].replace({'OLB': 'LB', 'ILB': 'LB'})    

df.to_csv(output)
df.head()
```
